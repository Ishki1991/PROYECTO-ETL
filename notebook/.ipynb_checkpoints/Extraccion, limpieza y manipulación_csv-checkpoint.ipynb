# Extraccion, limpieza y manipulación de datos

```python
import pandas as pd
from datetime import datetime
```

```python
# Vamos a comparar la temperatura actual y de los ultimos 5 años de las siguientes ciudades
mi_ciudad = "Madrid"
ciudad = "Barcelona"
otra_ciudad = "Bilbao"
```

```python
dtypes = {'Month': 'str',
          'Day': 'str',
          'Year': 'str',
          'AvgTemperature': 'float'}

df_datos = pd.read_csv("../data/city_temperature.csv", dtype = dtypes)

df_datos
```

```python
df_datos.info()
```

```python
df_datos2 = df_datos[(df_datos.City == mi_ciudad) | (df_datos.City == ciudad ) | (df_datos.City == otra_ciudad )]
df_datos2
```

```python
df_datos2['Month']= df_datos2['Month'].str.zfill(2)
df_datos2['Day']= df_datos2['Day'].str.zfill(2)
```

```python
df_datos2
```

```python
# Creamos una nueva columna "fecha" para unir el dia, mes y año.
df_datos2["fecha"] = df_datos2["Year"].astype(str) + '-' + df_datos2["Month"].astype(str) + '-' + df_datos2["Day"].astype(str)
```

```python
df_datos2.head()
```

```python
# Ahora quiero filtrarr los ultimos 5 años
df_datos3 = df_datos2[(df_datos2.Year == "2018")|
                      (df_datos2.Year == "2019")|
                      (df_datos2.Year == "2020")|
                      (df_datos2.Year == "2021")|
                      (df_datos2.Year == "2022")|
                      (df_datos2.Year == "2023")]
df_datos3
```

```python
df_datos3 = df_datos3.drop(columns = ["State"])
```

```python
#DF inicial
df_datos3.groupby(df_datos3.Year).count()
```

```python
df_datos3.info()
```

```python
df_datos3.to_csv("../data/DataFrame_inicial.csv", index=False)
```
