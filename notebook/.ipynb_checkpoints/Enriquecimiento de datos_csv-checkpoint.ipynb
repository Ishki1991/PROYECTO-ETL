# Enriquecimiento y limpieza de datos


```python
mi_ciudad = "Madrid"
ciudad = "Barcelona"
otra_ciudad = "Bilbao"
```

```python
#  Ya tengo mi df inicial. Busco datos para enriquecerlo
from geopy.geocoders import Nominatim

# Initialize Nominatim API
geolocator = Nominatim(user_agent="MyApp")

# ciudad Barcelona
location_ciudad = geolocator.geocode(ciudad)
lat_ciudad = location_ciudad.latitude
long_ciudad = location_ciudad.longitude

# mi ciudad Madrid
location_mi_ciudad = geolocator.geocode(mi_ciudad)
lat_mi_ciudad = location_mi_ciudad.latitude
long_mi_ciudad = location_mi_ciudad.longitude

# otra ciudad Bilbao
location_otra_ciudad = geolocator.geocode(otra_ciudad)
lat_otra_ciudad = location_otra_ciudad.latitude
long_otra_ciudad = location_otra_ciudad.longitude


print("Latitud de ", mi_ciudad, ":" ,lat_mi_ciudad)
print("Longitud de ", mi_ciudad, ":" ,long_mi_ciudad)
print("Latitud de ", ciudad,":" ,lat_ciudad)
print("Longitud de ", ciudad,":" ,long_ciudad)
print("Latitud de ", otra_ciudad,":" ,lat_otra_ciudad)
print("Longitud de ", otra_ciudad,":" ,long_otra_ciudad)
```

```python
from datetime import datetime
from meteostat import Point, Daily

# Set time period
start = datetime(2018, 1, 1)
end = datetime(2022, 12, 31)

#Datos Barcelona
location_ciudad = Point(lat_ciudad,long_ciudad)

data_ciudad = Daily(location_ciudad, start, end)
data_ciudad = data_ciudad.fetch().reset_index()

#Datos Madrid
location_mi_ciudad = Point(lat_mi_ciudad,long_mi_ciudad)

data_mi_ciudad = Daily(location_mi_ciudad, start, end)
data_mi_ciudad = data_mi_ciudad.fetch().reset_index()

#Datos Bilbao
location_otra_ciudad = Point(lat_otra_ciudad,long_otra_ciudad)

data_otra_ciudad = Daily(location_otra_ciudad, start, end)
data_otra_ciudad = data_otra_ciudad.fetch().reset_index()
```

```python
data_mi_ciudad = data_mi_ciudad.loc[:,["time","tmin","tmax","prcp","wspd","pres"]]

data_ciudad = data_ciudad.loc[:,["time","tmin","tmax","prcp","wspd","pres"]]

data_otra_ciudad = data_otra_ciudad.loc[:,["time","tmin","tmax","prcp","wspd","pres"]]

#crear una columna llamada city 
data_mi_ciudad["City"] = mi_ciudad

data_ciudad["City"] = ciudad

data_otra_ciudad["City"] = otra_ciudad
```

```python
#Cambiamos el nombre a la columna y luego el tipo para cruzar datos mas adelante
data_mi_ciudad = data_mi_ciudad.rename(columns = {"time": "fecha"})
data_ciudad = data_ciudad.rename(columns = {"time": "fecha"})
data_otra_ciudad = data_otra_ciudad.rename(columns = {"time": "fecha"})

data_ciudad["fecha"] = data_ciudad["fecha"].astype(str)
data_mi_ciudad["fecha"] = data_mi_ciudad["fecha"].astype(str)
data_otra_ciudad["fecha"] = data_otra_ciudad["fecha"].astype(str)
```

```python
# Madrid
data_mi_ciudad.groupby(data_ciudad.fecha.str[0:4]).count()
```

```python
# Barcelona
data_ciudad.groupby(data_ciudad.fecha.str[0:4]).count()
```

```python
# Bilbao
data_otra_ciudad.groupby(data_ciudad.fecha.str[0:4]).count()
```

```python
import pandas as pd
data_ciudades = pd.concat([data_ciudad, data_mi_ciudad, data_otra_ciudad])
data_ciudades
```

```python
cols = ["City","latitud","longitud"]
data = [["Barcelona",lat_ciudad, long_ciudad],
        ["Bilbao",lat_otra_ciudad,long_otra_ciudad],
        ["Madrid",lat_mi_ciudad,long_mi_ciudad]]

df_lat_lon = pd.DataFrame(data,columns=cols)
df_lat_lon
```

```python
data_ciudades1 = pd.merge(data_ciudades, df_lat_lon,  
                        how='left', 
                        left_on=["City"], 
                        right_on = ["City"])
data_ciudades1
```

```python
# Cruzo con mi df inicial (csv)
dtypes = {'Month': 'str',
          'Day': 'str',
          'Year': 'str',
          'AvgTemperature': 'float'}

df_datos = pd.read_csv("../data/DataFrame_inicial.csv", dtype = dtypes)
df_datos
```

```python
df_datos.info()
```

```python
df_datos_tot = pd.merge(df_datos, data_ciudades1,  
                        how='left', 
                        left_on=["fecha","City"], 
                        right_on = ["fecha","City"],)
df_datos_tot
```

```python
df_datos_tot.groupby([df_datos_tot.Year,df_datos_tot.City]).count()
```

```python
# Con eso hemos comprobado que no hay duplicados, aunque hay algunos nulos, los rellenamos con 0
df_datos_tot = df_datos_tot.fillna(0)
df_datos_tot.groupby([df_datos_tot.Year,df_datos_tot.City]).count() 
```

```python
df_datos_tot.to_csv("../data/DF_Final.csv", index=False)
```
