```python
mi_ciudad = "Madrid"
ciudad = "Barcelona"
otra_ciudad = "Bilbao"

import sys
sys.path.append("../")

import src.soporte as sp
```

```python
import pandas as pd
dtypes = {'Month': 'str',
          'Day': 'str',
          'Year': 'str'}

df_datos = pd.read_csv("../data/DF_Final.csv", dtype = dtypes)

df_datos
```

```python
# Ahora a extraer datos de inicio de año desde el año 2018 hasta el 2022

mes_str = str(1).zfill(2)
dia_str = str(1).zfill(2)

df_datos_grafico = df_datos[(df_datos.Month==mes_str)&(df_datos.Day==dia_str)]
df_datos_grafico
```

```python
fechas = df_datos_grafico['fecha'].tolist()
fechas = sorted(set(fechas))
print(fechas)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

fechas=df_datos_grafico['fecha'].tolist()
fechas=sorted(set(fechas))

avgtemp_mi_ciudad = df_datos_grafico[df_datos_grafico.City == mi_ciudad]['AvgTemperature'].tolist()
avgtemp_ciudad = df_datos_grafico[df_datos_grafico.City == ciudad]['AvgTemperature'].tolist()
avgtemp_otra_ciudad = df_datos_grafico[df_datos_grafico.City == otra_ciudad]['AvgTemperature'].tolist()

tmin_mi_ciudad = df_datos_grafico[df_datos_grafico.City == mi_ciudad]['tmin'].tolist()
tmin_ciudad = df_datos_grafico[df_datos_grafico.City == ciudad]['tmin'].tolist()
tmin_otra_ciudad = df_datos_grafico[df_datos_grafico.City == otra_ciudad]['tmin'].tolist()

tmax_mi_ciudad = df_datos_grafico[df_datos_grafico.City == mi_ciudad]['tmax'].tolist()
tmax_ciudad = df_datos_grafico[df_datos_grafico.City == ciudad]['tmax'].tolist()
tmax_otra_ciudad = df_datos_grafico[df_datos_grafico.City == otra_ciudad]['tmax'].tolist()

prcp_mi_ciudad = df_datos_grafico[df_datos_grafico.City == mi_ciudad]['prcp'].tolist()
prcp_ciudad=df_datos_grafico[df_datos_grafico.City == ciudad]['prcp'].tolist()
prcp_otra_ciudad=df_datos_grafico[df_datos_grafico.City == otra_ciudad]['prcp'].tolist()

wspd_mi_ciudad = df_datos_grafico[df_datos_grafico.City == mi_ciudad]['wspd'].tolist()
wspd_ciudad = df_datos_grafico[df_datos_grafico.City == ciudad]['wspd'].tolist()
wspd_otra_ciudad = df_datos_grafico[df_datos_grafico.City == otra_ciudad]['wspd'].tolist()

pres_mi_ciudad = df_datos_grafico[df_datos_grafico.City == mi_ciudad]['pres'].tolist()
pres_ciudad = df_datos_grafico[df_datos_grafico.City == ciudad]['pres'].tolist()
pres_otra_ciudad = df_datos_grafico[df_datos_grafico.City == otra_ciudad]['pres'].tolist()


#Imprimo los datos historicos
print('\n Mostramos algunos datos climatológicos históricos:')
fig = plt.figure(figsize=(8.5,2))
plt.grid()
plt.plot(fechas,avgtemp_mi_ciudad,label='Temp media '+mi_ciudad);
plt.plot(fechas,avgtemp_ciudad,label='Temp media '+ciudad);
plt.plot(fechas,avgtemp_otra_ciudad,label='Temp media '+otra_ciudad);
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.show()

fig = plt.figure(figsize=(8.5,2))
plt.grid()
plt.plot(fechas,tmin_mi_ciudad,label='Temp min '+mi_ciudad);
plt.plot(fechas,tmin_ciudad,label='Temp min '+ciudad);
plt.plot(fechas,tmin_otra_ciudad,label='Temp min '+otra_ciudad);
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.show()

fig = plt.figure(figsize=(8.5,2))
plt.grid()
plt.plot(fechas,tmax_mi_ciudad,label='Temp max '+mi_ciudad);
plt.plot(fechas,tmax_ciudad,label='Temp max '+ciudad);
plt.plot(fechas,tmax_otra_ciudad,label='Temp max '+otra_ciudad);
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.show()

fig = plt.figure(figsize=(8.5,2))
plt.grid()
plt.plot(fechas,prcp_mi_ciudad,label='Precipitaciones '+mi_ciudad);
plt.plot(fechas,prcp_ciudad,label='Precipitaciones '+ciudad);
plt.plot(fechas,prcp_otra_ciudad,label='Precipitaciones '+otra_ciudad);
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.show()

fig = plt.figure(figsize=(8.5,2))
plt.grid()
plt.plot(fechas,wspd_mi_ciudad,label='Velocidad del viento '+mi_ciudad);
plt.plot(fechas,wspd_ciudad,label='Velocidad del viento '+ciudad);
plt.plot(fechas,wspd_otra_ciudad,label='Velocidad del viento '+otra_ciudad);
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.show()

fig = plt.figure(figsize=(8.5,2))
plt.grid()
plt.plot(fechas,pres_mi_ciudad,label='Presion atm '+mi_ciudad);
plt.plot(fechas,pres_ciudad,label='Presion atm '+ciudad);
plt.plot(fechas,pres_otra_ciudad,label='Presion atm '+otra_ciudad);
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.show()
```

Segun la tabla y los gráficos, la presion atemosferica en las ciudades de Barcelona y Bilbao son relativamente iguales.
